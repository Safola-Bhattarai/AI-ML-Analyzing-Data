## Task-1: Unit Conversion Program

```python
def convert_units(choice, value):
    if choice == 1:
        return value * 3.28084, "Meters to Feet", "ft"
    elif choice == 2:
        return value / 3.28084, "Feet to Meters", "m"
    elif choice == 3:
        return value * 2.20462, "Kilograms to Pounds", "lbs"
    elif choice == 4:
        return value / 2.20462, "Pounds to Kilograms", "kg"
    elif choice == 5:
        return value * 0.264172, "Liters to Gallons", "gal"
    elif choice == 6:
        return value / 0.264172, "Gallons to Liters", "L"
    else:
        return None, None, None


print("Unit Converter")
print("1. Meters to Feet")
print("2. Feet to Meters")
print("3. Kilograms to Pounds")
print("4. Pounds to Kilograms")
print("5. Liters to Gallons")
print("6. Gallons to Liters")

try:
    choice = int(input("Enter your choice (1-6): "))
    value = float(input("Enter the value to convert: "))

    result, conversion_name, unit = convert_units(choice, value)

    if result is not None:
        print(conversion_name + " = " + str(round(result, 2)) + " " + unit)
    else:
        print("Invalid conversion choice.")

except ValueError:
    print("Error: Please enter numeric values only.")
```

### Task-2: Math Operations on a List of Numbers

```python

def find_sum(numbers):
    return sum(numbers)
def find_average(numbers):
    return sum(numbers) / len(numbers)
def find_maximum(numbers):
    return max(numbers)
def find_minimum(numbers):
    return min(numbers)

print("Mathematical Operations Program")
print("1. Find Sum")
print("2. Find Average")
print("3. Find Maximum")
print("4. Find Minimum")
try:
    choice = int(input("Enter your choice (1-4): "))
    user_input = input("Enter numbers separated by spaces: ")
    numbers = [float(num) for num in user_input.split()]

    if len(numbers) == 0:
        print("Error: List is empty.")
    else:
        if choice == 1:
            print("Sum =", find_sum(numbers))

        elif choice == 2:
            print("Average =", find_average(numbers))

        elif choice == 3:
            print("Maximum =", find_maximum(numbers))

        elif choice == 4:
            print("Minimum =", find_minimum(numbers))

        else:
            print("Invalid choice.")
except ValueError:
    print("Error: Please enter valid numeric values.")
```

## 3.2 Exercise on List Manipulation

#### Extract Every Other Element, slice a sublist, reverse a list using slicing, remove the first and last elements, get the first n elements, extract eleemnts from the end, extract elements in reverse order

```python
# Extract every other element
def extract_every_other(lst):
    return lst[::2]

# Return sublist from indexes
def get_sublist(lst, start, end):
    return lst[start:end+1]

# Reverse list using slicing
def reverse_list(lst):
    return lst[::-1]

# Remove first and last
def remove_first_last(lst):
    return lst[1:-1]

# Get first n elements
def get_first_n(lst, n):
    return lst[:n]

# Get last n elements
def get_last_n(lst, n):
    return lst[-n:]

# Reverse skipping one element
def reverse_skip(lst):
    return lst[-2::-2]

# Example list
numbers = [1, 2, 3, 4, 5, 6]
print("Original List:", numbers)
print("Every Other Element:", extract_every_other(numbers))
print("Sublist (2 to 4):", get_sublist(numbers, 2, 4))
print("Reversed List:", reverse_list(numbers))
print("Without First and Last:", remove_first_last(numbers))
print("First 3 Elements:", get_first_n(numbers, 3))
print("Last 2 Elements:", get_last_n(numbers, 2))
print("Reverse Skip List:", reverse_skip(numbers))
```

## 3.3 Exercise on Nested Lists

#### 1. Flatten a Nested List

```python
def flatten(lst):
    result = []

    for item in lst:
        for value in item:
            result.append(value)

    return result


print(flatten([[1, 2], [3, 4], [5]]))
```

#### 2. Accessing Nested List Elements

```python
def access_nested_element(lst, indices):
    return lst[indices[0]][indices[1]]


lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]

print(access_nested_element(lst, indices))
```

#### 3. Sum of All Elements in a Nested List

```python
def sum_nested(lst):
    total = 0

    for item in lst:
        if type(item) == list:
            total += sum_nested(item)
        else:
            total += item

    return total


print(sum_nested([[1, 2], [3, [4, 5]], 6]))
```

#### 4. Remove Specific Element from a Nested List

```python
def remove_element(lst, elem):
    result = []

    for item in lst:
        new_list = []

        for value in item:
            if value != elem:
                new_list.append(value)

        result.append(new_list)

    return result


lst = [[1, 2], [3, 2], [4, 5]]

print(remove_element(lst, 2))
```

#### 5. Find the Maximum Element in a Nested List

```python
def find_max(lst):
    biggest = 0

    for item in lst:
        if type(item) == list:
            value = find_max(item)

            if value > biggest:
                biggest = value

        else:
            if item > biggest:
                biggest = item

    return biggest


print(find_max([[1, 2], [3, [4, 5]], 6]))
```

#### 6. Count Occurrences of an Element in a Nested List

```python
def count_occurrences(lst, elem):
    count = 0

    for item in lst:
        for value in item:
            if value == elem:
                count += 1

    return count


lst = [[1, 2], [2, 3], [2, 4]]

print(count_occurrences(lst, 2))
```

#### 7. Deep Flatten (Any Depth)

```python
def deep_flatten(lst):
    result = []

    for item in lst:
        if type(item) == list:
            result += deep_flatten(item)
        else:
            result.append(item)

    return result


lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

print(deep_flatten(lst))
```

#### 8. Nested List Average

```python
def average_nested(lst):
    total = 0
    count = 0

    for item in lst:
        for value in item:
            total += value
            count += 1

    return total / count


print(average_nested([[1, 2], [3, 4], [5, 6]]))
```

#### Problem 4 Matrix Operation

```python
import numpy as np

# Given matrices
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

# 1. Prove A * A^-1 = I

A_inv = np.linalg.inv(A)

print("A inverse:")
print(A_inv)

print("A * A^-1:")
print(np.dot(A, A_inv))


# 2. Prove AB != BA

AB = np.dot(A, B)
BA = np.dot(B, A)

print("AB:")
print(AB)

print("BA:")
print(BA)


# 3. Prove (AB)^T = B^T A^T

left_side = (AB).T

right_side = np.dot(B.T, A.T)

print("(AB)^T:")
print(left_side)

print("B^T A^T:")
print(right_side)
```

#### Solving Linear equation using inverse method 

```python
import numpy as np

# Matrix A
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

# Matrix B
B = np.array([[-1],
              [-3],
              [9]])

# Find inverse of A
A_inv = np.linalg.inv(A)

# Solve X = A^-1 * B
X = np.dot(A_inv, B)

print("Solution:")
print(X)
```
